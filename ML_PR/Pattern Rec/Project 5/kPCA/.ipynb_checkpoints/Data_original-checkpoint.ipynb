```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine, load_digits

# Load Wine dataset
wine_data = load_wine()

# Create a DataFrame to hold the data
wine_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

# Add target column to DataFrame
wine_df['target'] = wine_data.target

# Map feature names to their descriptions
feature_description_map = {
    'alcohol': 'Alcohol content',
    'malic_acid': 'Malic acid content',
    'ash': 'Ash content',
    'alcalinity_of_ash': 'Alcalinity of ash',
    'magnesium': 'Magnesium content',
    'total_phenols': 'Total phenols content',
    'flavanoids': 'Flavanoids content',
    'nonflavanoid_phenols': 'Nonflavanoid phenols content',
    'proanthocyanins': 'Proanthocyanins content',
    'color_intensity': 'Color intensity',
    'hue': 'Hue',
    'od280/od315_of_diluted_wines': 'OD280/OD315 of diluted wines',
    'proline': 'Proline content'
}

# Rename DataFrame columns with feature descriptions
wine_df.rename(columns=feature_description_map, inplace=True)

# Map target labels to their descriptions
target_labels = wine_data.target_names

# Rename target column with label descriptions
wine_df['target'] = wine_df['target'].map(lambda x: target_labels[x])

# Display the DataFrame
print("Original Wine Dataset:")
wine_df.head()
```

```python
# Load Digits dataset
digits_data = load_digits()

# Extract images and labels
images = digits_data.images
labels = digits_data.target

# Display a few sample images
num_images_to_display = 5
fig, axes = plt.subplots(1, num_images_to_display, figsize=(12, 4))

for i in range(num_images_to_display):
    axes[i].imshow(images[i], cmap='gray')
    axes[i].set_title(f"Label: {labels[i]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
